```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Загрузка датасетов
team_statistics = pd.read_csv('Статистика команд.csv', sep=',', encoding='utf-8')
match_results = pd.read_csv('hltv_match_results.csv', sep=',', encoding='utf-8')

# Создание словаря для быстрого поиска статистики по команде
team_stats_dict = team_statistics.set_index('team').to_dict(orient='index')

# Функция для вычисления разности статистик команд
def compute_stat_diff(row):
    team1_stats = team_stats_dict.get(row['Команда 1'], None)
    team2_stats = team_stats_dict.get(row['Команда 2'], None)

    if team1_stats is not None and team2_stats is not None:
        # Разность вектора статистик
        stat_diff = [team1_stats[col] - team2_stats[col] for col in team1_stats.keys()]
        return stat_diff
    else:
        return [0] * len(team1_stats)

# Применение функции для создания разницы статистик
match_results['stat_diff'] = match_results.apply(compute_stat_diff, axis=1)

# Преобразование разности статистик в DataFrame
X = pd.DataFrame(match_results['stat_diff'].tolist())

# Подготовка целевой переменной: 1 если победила команда 1, 0 если команда 2
y = match_results['Победитель'].apply(lambda x: 1 if x == match_results['Команда 1'] else 0).values

# Стандартизация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Преобразование в тензоры для PyTorch
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# Создание модели
class MatchPredictor(nn.Module):
    def __init__(self, input_size):
        super(MatchPredictor, self).__init__()
        self.fc1 = nn.Linear(input_size, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return self.sigmoid(x)

# Инициализация модели
model = MatchPredictor(input_size=X_train.shape[1])

# Определение функции потерь и оптимизатора
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Обучение модели
epochs = 100
for epoch in range(epochs):
    model.train()

    # Прямой проход
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    # Обратный проход и оптимизация
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# Оценка модели
model.eval()
with torch.no_grad():
    outputs_test = model(X_test_tensor)
    predicted = (outputs_test > 0.5).float()
    accuracy = (predicted == y_test_tensor).float().mean()
    print(f'Accuracy on test data: {accuracy:.4f}')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from torch.optim.lr_scheduler import ReduceLROnPlateau

# 1. Чтение данных из CSV файлов
stats_df = pd.read_csv('/content/Статистика команд.csv')  # Статистика команд
matches_df = pd.read_csv('/content/hltv_match_results.csv')  # Результаты матчей

# Преобразуем все столбцы статистики в числовые значения
cols_to_convert = ['maps played', 'wins', 'losses', 'total kills', 'total deaths', 'rounds played', 'k/d ratio']
stats_df[cols_to_convert] = stats_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Приводим все названия команд к верхнему регистру и удаляем лишние пробелы
stats_df['team'] = stats_df['team'].str.upper().str.strip()

# Маппинг команд в индексы статистики
team_to_index = {team: idx for idx, team in enumerate(stats_df['team'])}

# Функция для получения статистики команды по имени
def get_team_stats(team_name):
    team_name = team_name.upper().strip()  # Приводим название команды к верхнему регистру и убираем пробелы
    if team_name not in team_to_index:
        return None  # Возвращаем None, если команда не найдена
    idx = team_to_index[team_name]
    return stats_df.iloc[idx][1:].values  # Берем все, кроме имени команды

# 2. Подготовка данных
X = []
y = []

# Создаем разницу в статистике для каждого матча
for _, row in matches_df.iterrows():
    team1 = row['Команда 1']
    team2 = row['Команда 2']
    winner = row['Победитель']

    # Получаем статистику команд
    stats_team1 = get_team_stats(team1)
    stats_team2 = get_team_stats(team2)

    # Пропускаем матч, если хотя бы одна из команд не найдена в статистике
    if stats_team1 is None or stats_team2 is None:
        continue

    # Разница в статистике (team1 - team2)
    diff_stats = stats_team1 - stats_team2  # Вычитание векторов статистики
    X.append(diff_stats)

    # Целевая переменная: 1 если победила команда 1, 0 если победила команда 2
    if winner == team1:
        y.append(0)  # 0 для победы команды 1
    else:
        y.append(1)  # 1 для победы команды 2

X = np.array(X)
y = np.array(y)

# 3. Стандартизация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Разделение на тренировочные и тестовые данные
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Преобразуем данные в тензоры PyTorch
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.long)  # Изменение: теперь y - это метки классов (целые числа)
y_test = torch.tensor(y_test, dtype=torch.long)

# Модель с двумя выходами
class MultiClassPredictor(nn.Module):
    def __init__(self, input_size):
        super(MultiClassPredictor, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_size, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 2),  # 2 выходных нейрона для двух классов
        )

    def forward(self, x):
        return self.network(x)  # Здесь не применяем softmax, потому что CrossEntropyLoss применяет softmax

# Инициализация модели
model = MultiClassPredictor(input_size=X_train.shape[1])
criterion = nn.CrossEntropyLoss()  # Используем CrossEntropyLoss для многоклассовой классификации
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Добавление learning rate scheduler (уменьшение скорости обучения, если loss не улучшается)
scheduler = ReduceLROnPlateau(optimizer, 'min', patience=10, factor=0.5, verbose=True)

# Для ранней остановки (early stopping)
best_loss = np.inf
patience_counter = 0
early_stopping_patience = 20

# Обучение модели
num_epochs = 200
for epoch in range(num_epochs):
    model.train()

    optimizer.zero_grad()
    outputs = model(X_train)  # Получаем логиты для тренировки
    loss = criterion(outputs, y_train)  # Применяем CrossEntropyLoss

    loss.backward()
    optimizer.step()

    # Проверка на улучшение модели
    if loss.item() < best_loss:
        best_loss = loss.item()
        patience_counter = 0  # Сбрасываем счетчик терпения
    else:
        patience_counter += 1

    # Обновление скорости обучения
    scheduler.step(loss.item())

    # Выводим информацию каждые 10 эпох
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

    # Ранняя остановка
    if patience_counter >= early_stopping_patience:
        print("Early stopping triggered!")
        break

# Оценка модели
model.eval()
with torch.no_grad():
    outputs = model(X_test)  # Получаем логиты для тестового набора

    # Печать логитов перед softmax
    print("Logits:", outputs[:5])  # Печать первых 5 логитов для анализа

    # Применяем softmax для получения вероятностей
    prob = torch.softmax(outputs, dim=1)

    # Печать вероятностей
    print("Softmax probabilities:", prob[:5])  # Печать первых 5 вероятностей

    predicted = torch.argmax(prob, dim=1)  # Получаем классы с максимальными вероятностями

    # Вычисляем метрики
    accuracy = (predicted == y_test).sum().item() / y_test.size(0)
    precision = precision_score(y_test.numpy(), predicted.numpy(), average='binary')
    recall = recall_score(y_test.numpy(), predicted.numpy(), average='binary')
    f1 = f1_score(y_test.numpy(), predicted.numpy(), average='binary')

    print(f'Accuracy: {accuracy * 100:.2f}%')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1-Score: {f1:.2f}')

    # Печать вероятностей победы для каждой команды
    prob_team1 = prob[:, 0].mean().item()
    prob_team2 = prob[:, 1].mean().item()
    print(f'Средняя вероятность победы первой команды: {prob_team1:.2f}')
    print(f'Средняя вероятность победы второй команды: {prob_team2:.2f}')
```

```python
import lightgbm as lgb
import pandas as pd
import numpy as np

def get_team_stats(team_name, stats_file_path):
    stats_df = pd.read_csv(stats_file_path)
    team_row = stats_df[stats_df['team'] == team_name]
    if team_row.empty:
        raise ValueError(f"Команда '{team_name}' не найдена в статистике.")
    return team_row.iloc[0, 1:].values  # Возвращаем статистику без имени команды


def predict_match_winner(team1, team2, model_path, stats_file_path):
    # Получаем статистику команд
    stats_team1 = get_team_stats(team1, stats_file_path)
    stats_team2 = get_team_stats(team2, stats_file_path)

    # Пропускаем матч, если хотя бы одна из команд не найдена
    if stats_team1 is None or stats_team2 is None:
        print("Одна или обе команды не найдены в базе данных.")
        return None, None

    # Загружаем модель LightGBM
    model = lgb.Booster(model_file=model_path)

    # Разница в статистике
    diff_stats = stats_team1 - stats_team2

    # Преобразуем разницу статистик в формат, подходящий для LightGBM
    diff_stats = np.array(diff_stats).reshape(1, -1)

    # Прогнозируем вероятности для двух классов
    prob = model.predict(diff_stats)  # Вероятности для двух классов

    prob_team1 = prob[0]  # Вероятность победы первой команды
    prob_team2 = 1 - prob_team1  # Вероятность победы второй команды (1 - prob_team1)

    return prob_team1, prob_team2  # Возвращаем вероятности для обеих команд


model_path = 'prediction_model_last_lgmb.txt'  # Путь к сохраненной модели LightGBM
stats_file_path = '/content/Статистика команд.csv'  # Путь к CSV файлу с данными статистики команд

# Пример команд
team1 = 'Vitality'
team2 = 'G2'

# Получаем вероятности победы
prob_team1, prob_team2 = predict_match_winner(team1, team2, model_path, stats_file_path)

if prob_team1 is not None and prob_team2 is not None:
    print(f"Вероятность победы {team1}: {prob_team1:.2f}")
    print(f"Вероятность победы {team2}: {prob_team2:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import lightgbm as lgb

# 1. Чтение данных из CSV файлов
stats_df = pd.read_csv('/content/Статистика команд.csv')  # Статистика команд
matches_df = pd.read_csv('/content/hltv_match_results.csv')  # Результаты матчей

# Преобразуем все столбцы статистики в числовые значения
cols_to_convert = ['maps played', 'wins', 'losses', 'total kills', 'total deaths', 'rounds played', 'k/d ratio']
stats_df[cols_to_convert] = stats_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Приводим все названия команд к верхнему регистру и удаляем лишние пробелы
stats_df['team'] = stats_df['team'].str.upper().str.strip()

# Маппинг команд в индексы статистики
team_to_index = {team: idx for idx, team in enumerate(stats_df['team'])}

# Функция для получения статистики команды по имени
def get_team_stats(team_name):
    team_name = team_name.upper().strip()  # Приводим название команды к верхнему регистру и убираем пробелы
    if team_name not in team_to_index:
        return None  # Возвращаем None, если команда не найдена
    idx = team_to_index[team_name]
    return stats_df.iloc[idx][1:].values  # Берем все, кроме имени команды

# 2. Подготовка данных
X = []
y = []

# Создаем разницу в статистике для каждого матча
for _, row in matches_df.iterrows():
    team1 = row['Команда 1']
    team2 = row['Команда 2']
    winner = row['Победитель']

    # Получаем статистику команд
    stats_team1 = get_team_stats(team1)
    stats_team2 = get_team_stats(team2)

    # Пропускаем матч, если хотя бы одна из команд не найдена в статистике
    if stats_team1 is None or stats_team2 is None:
        continue

    # Разница в статистике (team1 - team2)
    diff_stats = stats_team1 - stats_team2  # Вычитание векторов статистики
    X.append(diff_stats)

    # Целевая переменная: 1 если победила команда 1, 0 если победила команда 2
    if winner == team1:
        y.append(0)  # 0 для победы команды 1
    else:
        y.append(1)  # 1 для победы команды 2

X = np.array(X)
y = np.array(y)

# 3. Стандартизация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Разделение на тренировочные и тестовые данные
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 5. Обучение модели LightGBM
lgb_train = lgb.Dataset(X_train, y_train)
lgb_test = lgb.Dataset(X_test, y_test, reference=lgb_train)

params = {
    'objective': 'binary',  # Бинарная классификация
    'metric': 'binary_logloss',  # Логарифмическая функция потерь
    'boosting_type': 'gbdt',  # Градиентный бустинг
    'learning_rate': 0.01,
    'num_leaves': 31,
    'max_depth': -1,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': -1
}

# Функция для ранней остановки
early_stopping_callback = lgb.early_stopping(stopping_rounds=50)

# Словарь для сохранения результатов валидации
eval_result = {}

# Тренировка модели с ранней остановкой и сохранением метрик
model = lgb.train(
    params,
    lgb_train,
    num_boost_round=500,
    valid_sets=[lgb_train, lgb_test],
    valid_names=['train', 'eval'],
    callbacks=[early_stopping_callback, lgb.record_evaluation(eval_result)]
)

# 6. Оценка модели
y_pred_proba = model.predict(X_test, num_iteration=model.best_iteration)  # Вероятности
y_pred = (y_pred_proba > 0.5).astype(int)  # Прогнозируемые классы

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')
print("Confusion Matrix:")
print(conf_matrix)

# Средние вероятности побед для каждой команды
prob_team1_avg = y_pred_proba.mean()  # Средняя вероятность победы первой команды
prob_team2_avg = 1 - prob_team1_avg   # Средняя вероятность победы второй команды
print(f'Средняя вероятность победы первой команды: {prob_team1_avg:.2f}')
print(f'Средняя вероятность победы второй команды: {prob_team2_avg:.2f}')
```
